```python
mamdani_r = requests.get("https://api.elections.kalshi.com/trade-api/v2/markets/KXMAYORNYCPARTY-25-D")
mamdani_orderbook = mamdani_r.json()
mamdani_yes_bid = mamdani_orderbook["market"]["yes_bid"]
mamdani_yes_ask = mamdani_orderbook["market"]["yes_ask"]
mamdani_no_bid = mamdani_orderbook["market"]["no_bid"]
mamdani_no_ask = mamdani_orderbook["market"]["no_ask"]
print(f"Yes Bid: {mamdani_yes_bid}\n \
        Yes Ask: {mamdani_yes_ask}\n \
        No Bid: {mamdani_no_bid}\n \
        No Ask: {mamdani_no_ask}\n")
```
